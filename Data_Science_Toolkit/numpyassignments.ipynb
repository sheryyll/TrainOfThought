### Assignment 1: Array Creation and Manipulation

1. Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.
2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.

```python
# 1. Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.
import numpy as np
import random
arr1 = np.random.randint(1,21, size=(5,5))
print(f'Array before modification: \n{arr1}\n')

arr1[:, 2] = 1
print(f'Array after modification: \n{arr1}')
```

```python
# 2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.
arr2 = np.random.randint(1, 17, size=(4,4))
print(f'Array before modification : \n{arr2}\n')
np.fill_diagonal(arr2, 0)
print(f'Array after modification: \n{arr2}')
```

### Assignment 2: Array Indexing and Slicing

1. Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.
2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
# 1
arr3 = np.random.randint(1, 37, size=(6,6))
print(f'Original Array: \n{arr3}\n')
sub_arr3 = arr3[3:6, 2:5]
print(f'Sub array: \n{sub_arr3}')
```

```python
# 2 
arr4 = np.random.randint(1, 50, size=(5,5))
print(f'Original Array: \n{arr4}\n')

top = arr4[0, :]
bottom = arr4[-1, :]
left = arr4[1:-1, 0]
right = arr4[1:-1, -1]

border_elements = np.concatenate([top, right, bottom[::-1], left[::-1]])

print("\nBorder elements:")
print(border_elements)
```

### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.
2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.


```python
#1
arr5_a = np.random.randint(1,50 , size=(3,4))
arr5_b = np.random.randint(1,50, size=(3,4))
print(f'Array A: \n{arr5_a}\n')
print(f'Array B: \n{arr5_b}\n')

sum_arr5 = arr5_a + arr5_b
print(f'Element wise sum of two arrays: \n{sum_arr5}\n')

diff_arr5 = arr5_a - arr5_b
print(f'Element wise difference of two arrays: \n{diff_arr5}\n')

mul_arr5 = arr5_a * arr5_b
print(f'Element wise multiplication of two arrays: \n{mul_arr5}\n')

div_arr5 = arr5_a // arr5_b
print(f'Element wise division of two arrays: \n{div_arr5}\n')
```

```python
#2 
arr6 = np.random.randint(1, 17, size=(4,4))
print(f'Original Array: \n{arr6}\n')

row_sum = np.sum(arr6, axis = 1)
col_sum = np.sum(arr6, axis = 0)
print(f'Sum of each row: \n{row_sum}\n')
print(f'Sum of each column: \n{col_sum}\n')
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
# 1
arr7 = np.random.randint(1, 50, size=(5,5))
print(f'Original Array: \n{arr7}\n')
print(f'Mean: {np.mean(arr7)}')
print(f'Median: {np.median(arr7)}')
print(f'Standard Deviation: {np.std(arr7)}')
print(f'Variance: {np.var(arr7)}')
```

```python
# 2
arr8 = np.random.randint(1, 10, size=(3,3))
print(f'Original Array: \n{arr8}\n')
mean_arr8 = np.mean(arr8)
std_arr8 = np.std(arr8)
normalized_arr8 = (arr8 - mean_arr8) / std_arr8
print(f'Normalized Array: \n{normalized_arr8}\n')
```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

```python

arr11 = np.random.randint(1, 26, size=(3,3))
print(f"Original matrix:\n{arr11}\n")

determinant = np.linalg.det(arr11)
print(f"Determinant:\n{determinant}\n")

inverse = np.linalg.inv(arr11)
print(f"Inverse:\n{inverse}\n")

eigenvalues = np.linalg.eigvals(arr11)
print(f"Eigenvalues:\n{eigenvalues}")
```

```python
arr12_a = np.random.randint(1, 26, size=(2,3))
arr12_b = np.random.randint(1, 26, size=(3,2))
print(f"Original matrices are:\n{arr12_a}\n{arr12_b}\n")

arr12_c = np.dot(arr12_a, arr12_b)
print(f"Multiplication result:\n{arr12_c}")
```

### Assignment 7: Advanced Array Manipulation

1. Create a NumPy array of shape (3, 3) with values from 1 to 9. Reshape the array to shape (1, 9) and then to shape (9, 1).
2. Create a NumPy array of shape (5, 5) filled with random integers. Flatten the array and then reshape it back to (5, 5).

```python
# 1
arr13 = np.random.randint(1, 10, size=(3,3))
print(f"Original matrix:\n{arr13}\n")

reshaped1 = arr13.reshape((1,9))
print(f"Reshaped: \n{reshaped1}\n")

reshaped2 = arr13.reshape((9,1))
print(f"Reshaped: \n{reshaped2}\n")
```

```python
arr14 = np.random.randint(1, 26, size=(5,5))
print(f"Original matrix:\n{arr14}\n")

flattened = arr14.flatten()
print(f"Flattened array:\n{flattened}\n")

reshape1 = flattened.reshape((5,5))
print(f"Reshaped array: \n{reshape1}")
```

### Assignment 8: Fancy Indexing and Boolean Indexing

1. Create a NumPy array of shape (5, 5) filled with random integers. Use fancy indexing to extract the elements at the corners of the array.
2. Create a NumPy array of shape (4, 4) filled with random integers. Use boolean indexing to set all elements greater than 10 to 10.

```python
arr15 = np.random.randint(1, 26, size=(5,5))
print(f"Original matrix:\n{arr15}\n")
corners = arr15[[0, 0, -1, -1], [0, -1, 0, -1]]
print(f"Corner elements:\n{corners}")
```

```python
arr16 = np.random.randint(1, 26, size=(4,4))
print(f"Original matrix:\n{arr15}\n")

arr16[arr16 > 10] = 10
print(f"Modified array:\n{arr16}")
```
