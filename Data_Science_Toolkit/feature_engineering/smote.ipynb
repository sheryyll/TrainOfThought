## SMOTE (Synthetic Minority Over-sampling Technique)

SMOTE is used to handle imbalanced datasets by creating new synthetic samples for the minority class.  
It generates new data points by interpolating between existing minority samples.


```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, 
    weights=[0.9], n_redundant=0, random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(X, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])

final_df = pd.concat([df1,  df2], axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#  tranforming dataset using SMOTE
oversample = SMOTE()
X,y = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```
