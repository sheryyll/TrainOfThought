```python
# Import libraries
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
# Sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
sent
```

```python
# Vocabulary size
voc_size = 10000
```

```python
# One-hot encoding
one_hot_repr = [one_hot(words, voc_size) for words in sent]
print("One-hot representation:\n", one_hot_repr)
```

```python
# Padding
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
print("\nPadded sequences:\n", embedded_docs)
```

```python
# Embedding model
dim = 10
```

```python
model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
# Get embeddings
output = model.predict(embedded_docs)
print("\nEmbedding output shape:", output.shape)
```
