```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
df = pd.DataFrame(dataset.data, columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's23', 's4', 's5', 's6'])
df.head()
```

```python
X = df
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
# correalation
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train) 
```

#### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
param_grid = {
    'DecisionTree__criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'DecisionTree__splitter': ['best', 'random'],
    'DecisionTree__max_depth': np.arange(1,26),
    'DecisionTree__max_features': [None, 'sqrt', 'log2']
}
```

```python
pipeline = Pipeline([
    ('DecisionTree', DecisionTreeRegressor())
])
```

```python
grid = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv= 5, scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2 score:", r2_score(y_test, y_pred))
print("\nMean absolute error : ", mean_absolute_error(y_test, y_pred))
print("\nMean sqaured error : ", mean_absolute_error(y_test, y_pred))
```

```python
selected_model = DecisionTreeRegressor(criterion='friedman_mse', max_depth=2, max_features=None, splitter='random')
```

```python
selected_model.fit(X_train, y_train)
```

```python
# visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected_model, filled=True)
plt.show()
```
