## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# creation of synthetic data points
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, 
                           n_clusters_per_class=2, n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1],hue=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
# prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
score = accuracy_score(y_test, y_pred)
print(f"Accuracy score : {score}")
print(f"Classification report: \n {classification_report(y_test, y_pred)}\n")
print(f"Confusion Matrix : \n {confusion_matrix(y_test, y_pred)}\n")
```

```python
svc.coef_
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train, y_train)
```

```python
# prediction
y_pred_1 = rbf.predict(X_test)
```

```python
score_1 = accuracy_score(y_test, y_pred_1)
print(f"Accuracy score : {score_1}")
print(f"Classification report: \n {classification_report(y_test, y_pred_1)}\n")
print(f"Confusion Matrix : \n {confusion_matrix(y_test, y_pred_1)}\n")
```

```python
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)
# prediction
y_pred_2 = poly.predict(X_test)
score_2 = accuracy_score(y_test, y_pred_2)
print(f"Accuracy score : {score_2}")
print(f"Classification report: \n {classification_report(y_test, y_pred_2)}\n")
print(f"Confusion Matrix : \n {confusion_matrix(y_test, y_pred_2)}\n")
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)
# prediction
y_pred_3 = sigmoid.predict(X_test)
score_3 = accuracy_score(y_test, y_pred_3)
print(f"Accuracy score : {score_3}")
print(f"Classification report: \n {classification_report(y_test, y_pred_3)}\n")
print(f"Confusion Matrix : \n {confusion_matrix(y_test, y_pred_3)}\n")
```

```python
sigmoid.intercept_
```

### Hyperparameter tuining

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, cv=5, verbose=3, n_jobs=1, refit=True)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_4 = grid.predict(X_test)
score_4 = accuracy_score(y_test, y_pred_4)
print(f"Accuracy score : {score_4}")
print(f"Classification report: \n {classification_report(y_test, y_pred_4)}\n")
print(f"Confusion Matrix : \n {confusion_matrix(y_test, y_pred_4)}\n")
```
