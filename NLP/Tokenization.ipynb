```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus = '''Ravi is a college student who studies Artificial Intelligence. 
He enjoys coding in Python and building machine learning models. 
In his free time, he plays chess and reads books. 

'''
```

```python
print(corpus)
```

```python
# Converting paragraph into  sentences
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
# Sentence -> Word & Paragraph -> word
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
# the above treats puncuations as a spearate word as well
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
# this doesnt treat '.' as a separate word only last fullstop is considered as a separate word
```
